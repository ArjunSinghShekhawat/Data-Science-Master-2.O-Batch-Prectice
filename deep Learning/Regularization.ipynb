```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=100,noise=0.25,random_state=42)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
model = Sequential()
```

```python
model.add(Dense(128,activation='relu',input_dim=2))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
adam = Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x,y,epochs=2000,validation_split=0.20,verbose=0)
```

```python
plot_decision_regions(x,y.astype('int'),clf=model,legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()
```

```python
model.add(Dense(128,activation='relu',input_dim=2,kernel_regularizer=tensorflow.keras.regularizers.l2(0.03)))
model.add(Dense(128,activation='relu',input_dim=2,kernel_regularizer=tensorflow.keras.regularizers.l2(0.03)))
model.add(Dense(1,activation='sigmoid'))
```

```python
adam = Adam(learning_rate=0.01)
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x,y,epochs=2000,validation_split=0.20,verbose=0)
```

```python
plot_decision_regions(x,y.astype('int'),clf=model,legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=500,noise=0.25,random_state=42)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
model1 = Sequential()
```

```python
model1.add(Dense(128,activation='relu',input_dim=2))
model1.add(Dense(128,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.summary()
```

```python
adam = Adam(learning_rate=0.01)
model1.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
history1  = model1.fit(x,y,epochs=1000,validation_split=0.2,verbose=1)
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
model2 = Sequential()
model2.add(Dense(128,activation='relu',input_dim=2,kernel_regularizer=tensorflow.keras.regularizers.l2(0.3)))
model2.add(Dense(128,activation='relu',kernel_regularizer=tensorflow.keras.regularizers.l2(0.3)))
model2.add(Dense(1,activation='sigmoid'))

adam = Adam(learning_rate=0.01)
model2.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
history2  = model2.fit(x,y,epochs=1000,validation_split=0.2,verbose=1)

plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
```

```python
plot_decision_regions(x,y.astype('int'),clf=model1,legend=2)
```

```python
plot_decision_regions(x,y.astype('int'),clf=model2,legend=2)
```
