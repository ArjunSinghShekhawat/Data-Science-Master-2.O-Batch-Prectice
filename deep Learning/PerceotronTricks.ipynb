```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
x,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=1,class_sep=10,hypercube=False)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,7))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=70)
```
