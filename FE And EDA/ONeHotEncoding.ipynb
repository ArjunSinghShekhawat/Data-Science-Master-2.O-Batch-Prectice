```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
df = pd.DataFrame({'color':['red','blue','blue','green','red','red','blue']})
```

```python
df
```

```python
encoder = OneHotEncoder()
```

```python
encoded_df = pd.DataFrame(encoder.fit_transform(df[['color']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
pd.concat([df,encoded_df],axis=1)
```
