```python
import requests
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as bs
import logging
```

```python
search = input("Enter searching key")
search = search.replace(" ","")
flibkart_url = "https://www.flipkart.com/search?q="+search
```

```python
flibkart_url
```

```python
client_url = uReq(flibkart_url)
```

```python
client_url
```

```python
flipkart_page = client_url.read()
```

```python
flipkart_html = bs(flipkart_page,'html.parser')
```
