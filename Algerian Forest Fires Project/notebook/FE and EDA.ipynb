```python
y = df['FWI']
x = df[['Temperature','RH','Ws','Rain','FFMC','DMC','DC','ISI','BUI','Classes','Region']]
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
x_train.corr()
```

```python
def correlation(dataset,threshold):
    corr_name = set()
    corr_matrix = dataset.corr()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]>threshold):
                colname = corr_matrix.columns[i]
                corr_name.add(colname)
    return corr_name
```

```python
feature = correlation(x_train,0.85)
```

```python
x_train.drop(feature,inplace=True,axis=1)
x_test.drop(feature,inplace=True,axis=1)
```

```python
x_train
```

```python
#model creation
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.preprocessing import StandardScaler

linear = LinearRegression()
scaler = StandardScaler()

x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
linear.fit(x_train_scale,y_train)
intercept = linear.intercept_
coff = linear.coef_
y_pred = linear.predict(x_test_scale)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_test,y_pred)
print(r2)

```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.preprocessing import StandardScaler

Ridge = Ridge()
scaler = StandardScaler()

x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
Ridge.fit(x_train_scale,y_train)
intercept = Ridge.intercept_
coff = Ridge.coef_
y_pred = Ridge.predict(x_test_scale)
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_test,y_pred)
print(r2)
```
