```python
gray = cv2.cvtColor(image1,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
gray
```

```python
#horizontal edge detection
filter_x = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
#vertical edge detection
filter_y = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])


filter_image_1 = cv2.filter2D(gray,-1,filter_x)
filter_image_2 = cv2.filter2D(gray,-1,filter_y)
f,(ax1,ax2) = plt.subplots(1,2,figsize=(20,10))
ax1.set_title('Filter X')
ax1.imshow(filter_image_1,cmap='gray')
ax2.set_title('Filter Y')
ax2.imshow(filter_image_2)
```
