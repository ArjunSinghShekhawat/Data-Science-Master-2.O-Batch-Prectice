```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.scatterplot(x=iris.sepal_width,y=iris.petal_width)
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_length)
```

```python
sns.histplot(iris['petal_length'])
```

```python
sns.barplot(x=iris['species'],y=iris['petal_length'])
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.relplot(x=tips['total_bill'],y=tips['tip'],hue=tips['day'])
```

```python
sns.relplot(x=tips['day'],y=tips['total_bill'],hue=tips['sex'])
```

```python
tips['sex'].value_counts()
```

```python
sns.pairplot(tips)
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.catplot(x=tips['day'],y=tips['total_bill'],hue=tips['sex'])
```
