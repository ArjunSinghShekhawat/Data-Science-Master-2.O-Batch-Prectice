Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'],hue=fmri['event'])
```

```python
fmri['event'].value_counts()
```

```python
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'],hue=fmri['region'])
```

```python
fmri['region'].value_counts()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
df = sns.load_dataset('diamonds')
```

```python
df.head()
```

```python
sns.histplot(data=df,x='price',hue='cut',multiple='stack')
```

```python
df['cut'].value_counts()
```

```python
sns.histplot(data=df,x='price',hue='cut')
```
