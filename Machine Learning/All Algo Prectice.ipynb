```python
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x,y = make_blobs(n_samples=1000,n_features=2,centers=3,random_state=999)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
x
```

```python
wcss = []
for k in range(1,20):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,20),wcss)
plt.xlabel('K value')
plt.ylabel('WCSS')
plt.show()
```

```python
from kneed import KneeLocator
```

```python
k = KneeLocator(range(1,20),wcss,direction='decreasing',curve='convex')
```

```python
k.elbow
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
```

```python
label = kmeans.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=label)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score_error = []

for k in range(2,20):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x)
    score = silhouette_score(x,kmeans.labels_)
    silhouette_score_error.append(score)
```

```python
silhouette_score_error
```

```python
plt.plot(range(2,20),silhouette_score_error)
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
x = data.data
```

```python
y = data.target
```

```python
x
```

```python
y
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
x_train = scale.fit_transform(x)
```

```python
pca = PCA(n_components=2)
```

```python
pca_scale = pca.fit_transform(x_train)
```

```python
pca_scale
```

```python
plt.scatter(pca_scale[:,0],pca_scale[:,1])
```

```python
import scipy.cluster.hierarchy as s
```

```python
plt.figure(figsize=(18,8))
plt.xlabel("Data Points")
plt.ylabel('Euclidean distance')
plt.title("Dendogram")
s.dendrogram(s.linkage(pca_scale,method='ward'))
plt.show()
```

```python
aglo = AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
```

```python
label = aglo.fit_predict(pca_scale)
```

```python
plt.scatter(pca_scale[:,0],pca_scale[:,1],c=label)
```

```python
print(silhouette_score(pca_scale,label))
```
