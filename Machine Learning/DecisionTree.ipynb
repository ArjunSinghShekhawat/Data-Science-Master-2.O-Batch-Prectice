```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
print(data.DESCR)
```

```python
data.data
```

```python
data.target
```

```python
d1 = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
d2 = pd.DataFrame(data.target,columns=['target'])
```

```python
df = pd.concat([d1,d2],axis=1)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(max_depth=2,criterion='entropy')
```

```python
classifier.fit(x_train,y_train)
```

```python
pre = classifier.predict(x_test)
```

```python
pre
```

```python
list(y_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(pre,y_test))
```

```python
print(accuracy_score(pre,y_test))
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(10,8))
tree.plot_tree(classifier,filled=True)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['target'] = data.target
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
df.columns
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
pca_scale = pca.fit_transform(x)
```

```python
pca_scale
```

```python
classifier = DecisionTreeClassifier()
```

```python
x_train,x_test,y_train,y_test = train_test_split(pca_scale,y,test_size=0.20,random_state=42)
```

```python
classifier.fit(x_train,y_train)
```

```python
pre = classifier.predict(x_test)
```

```python
pre
```

```python
confusion_matrix(pre,y_test)
```

```python
accuracy_score(pre,y_test)
```
