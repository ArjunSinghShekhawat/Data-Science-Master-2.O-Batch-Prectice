```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=1000,n_features=2,n_targets=1,noise=3.0)
```

```python
x
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
obj = SVR(kernel='linear')
```

```python
obj.fit(x_train,y_train)
```

```python
obj.intercept_
```

```python
obj.coef_
```

```python
y_pred = obj.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
a = r2_score(y_pred,y_test)
```

```python
1-(1-a)*(len(y_test)-1)/(len(y_test)-x_train.shape[1]-1)
```

```python
a
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {
    'C':np.linspace(1,3,5),
    'gamma':np.logspace(1,3,5),
    'kernel':['linear'],
    'epsilon':np.arange(1,2,0.2)
}
```

```python
clf = GridSearchCV(SVR(),param_grid=parameter,cv=5,refit=True,verbose=3)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
