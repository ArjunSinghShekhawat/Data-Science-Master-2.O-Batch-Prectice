```python
import pandas as pd
```

```python
from sklearn.datasets import make_circles
```

```python
x,y = make_circles(n_samples=1000,noise=0.1,random_state=42)
```

```python
import matplotlib.pylab as plt
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
obj = LocalOutlierFactor(n_neighbors=5,algorithm='auto')
```

```python
pre = obj.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=pre)
```

```python
pre
```

```python
from sklearn.ensemble import IsolationForest
```

```python
obj = IsolationForest(contamination='auto')
```

```python
da  = obj.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=da)
```
