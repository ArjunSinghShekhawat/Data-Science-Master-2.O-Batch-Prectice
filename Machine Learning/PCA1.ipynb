```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
x = data.data
y = data.target
```

```python
x
```

```python
y
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
pca_scale = pca.fit_transform(x)
```

```python
pca_scale
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
import plotly.express as p
```

```python
p.scatter_3d(pca_scale[:,0],pca_scale[:,1],pca_scale[:,2],color=y)
```

```python
pca_scale
```

```python
p.scatter_3d(pca_scale[:,0], pca_scale[:,1], pca_scale[:,2], color=y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_scale[:,0],pca_scale[:,1],pca_scale[:,2] c='b', marker='o')  # 'c' sets the color, 'marker' sets the marker style
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```
