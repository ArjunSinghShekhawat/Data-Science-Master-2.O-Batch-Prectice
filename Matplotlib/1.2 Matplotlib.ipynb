```python
import matplotlib.pyplot as plt
import numpy as np
```

Pie plot

```python
x = [50,20,15,5,10]
branch = ['Computer Science','Electrical','Civil','AI','Mechanical']
```

```python
c = ['red','yellow','brown','orange','skyblue']
plt.figure(figsize=(7,5))
plt.pie(x,labels=branch,colors=c,startangle=90,explode=[0.1,0.1,0.1,0.2,0.1],counterclock=True, shadow=True,autopct="%1.1f%%",pctdistance=0.5,labeldistance=1.1)
plt.legend(title="Engineering Branch")
plt.show()
```

Bar Chart

```python
plt.bar(branch,x,width=0.2,color='red')
```
