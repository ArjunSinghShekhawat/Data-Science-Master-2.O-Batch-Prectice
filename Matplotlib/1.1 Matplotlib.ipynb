```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,10,200)
```

```python
y = np.cos(x)
```

```python
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("y Axis")
plt.title("This is my cos graph")
```
