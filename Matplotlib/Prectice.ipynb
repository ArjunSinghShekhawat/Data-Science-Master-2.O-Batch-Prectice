```python
x = np.random.randint(50,100,200)
```

```python
y = np.log10(x)
```

```python
plt.plot(x,y)
```

```python
x = np.random.randn(50)
y = np.random.randn(50)
colours = np.random.randn(50)
size = 100*np.random.randn(50)

plt.scatter(x,y,c=colours,s=size,marker='v')
plt.show()
```

```python
x = ['a','b','c','d','e']
y = np.random.randint(1,100,5)
plt.bar(x,y)
```

```python
x = np.linspace(1,5,20)
y = np.linspace(1,5,20)
z = np.linspace(1,5,20)

fg = plt.figure(figsize=(4,5))
ax = fg.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```

```python
x = [10,55,5,15,7.5,7.5]
city = ['jaipur','sikar','churu','jodhpur','banwada',"chomu"]
plt.pie(x,labels=city,shadow=True,startangle=90,explode=[0.3,0,0,0,0,0],colors=['pink','yellow','red','blue','green','black'])
plt.legend(title="Arjun Pie Char")
plt.show()
```
